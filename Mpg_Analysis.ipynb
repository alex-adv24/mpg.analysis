#### #1. Create a visual showing the relationship between horsepower and fuel economy

##### Horsepower and mpg have a clear relationship. The vehicles with a high horsepower have a low fuel economy. And the vehicles with a low horsepower often have a high fuel economy.

```python
sns.relplot(data=mpg,
           x="mpg",y="horsepower",color='#5F1B')
plt.title('Horsepower & Fuel Economy')
plt.xlabel('MPG')
plt.ylabel('Horsepower')
plt.show()
```

#### #2.Create a table summarizing fuel economy by country of origin

##### This chart displays the average mpg of vehicles depending on their country/continent of origin. Japanese vehicles have the highest fuel economy followed by european vehicles. And American vehicles have the lowest mpg at an average of 20 MPG.

```python
average_mpg = mpg.groupby('origin')['mpg'].mean().reset_index()
average_mpg.columns = ['Country','Average_MPG']
average_mpg
```

#### #3.Create a visual showing how fuel economy has changed over time

##### Over time fuel economy in vehicles has gradually increased. In 1970 the average was about 17 MPG and in 1982 the average was about 31 MPG.

```python
sns.catplot(data=mpg,
           x="model_year",y="mpg",kind="bar",errorbar=None,color="#5F1B6B")
plt.title('Fuel Economy Change Over Time')
plt.xlabel('Model Year')
plt.ylabel('MPG')
plt.show()
```

#### #4.Create two other visuals or tables related to fuel economy

##### #4.A) Scatter plot graph showing the relationship between fuel economy, weight and horsepower. 
It is clear that horsepower and weight are closely linked. Vehicles that have high horsepower also have a high weight and vehicles with a low horsepower are typically lighter.

```python
sns.relplot(data=mpg,
           x="mpg",y="weight",hue='horsepower')
plt.title('Relationship of Fuel Economy with Weight & Horsepower')
plt.show()
```

##### #4.B) Bar chart displaying the top 5 Vehicles with the Highest Mpg
The top five vehicles with the highest mpg are all from Japan or Europe. This makes sense because vehicles made in either Japan or Europe have the best fuel economy while vehicles from America often have a low fuel economy.

```python
top5vehicles = mpg.nlargest(5,'mpg')
sns.barplot(data=top5vehicles,x='mpg',y='name',color='#C875C4')
plt.title('Vehicles with the Highest Mpg')
plt.ylabel('Vehicle Model')
plt.xlabel('MPG')
plt.show()
```
